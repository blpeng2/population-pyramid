```python
df_woman = pd.read_excel("./202211_202211_연령별인구현황_월간.xlsx", skiprows=3, usecols="B, AB:AV", index_col="행정기관")
df_woman
```

```python
df_woman.columns = df_man.columns
```

```python
df_woman
```

```python
df_woman.iloc[0] = df_woman.iloc[0].str.replace(",", "").astype(int)
```

```python
df_woman.iloc[0]
```

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams["font.family"] = "D2Coding" # Window
# matplotlib.rcParams["font.family"] = "AppleGothic" # Mac
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_man.columns, -df_man.iloc[0] // 1000) # 단위 : 천명
plt.barh(df_woman.columns, df_woman.iloc[0] // 1000)
plt.title("2022 population pyramid")
plt.savefig("2022 population pyramid.png", dpi=100)
```
